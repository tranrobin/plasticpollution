```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("plastics.csv")
```

```python
df.head(10)
```

```python
df = df[df['country'] != 'EMPTY']
df = df[df['parent_company'] != 'Unbranded']
df = df[df['parent_company'] != 'Grand Total']
df.dropna(subset = ['parent_company'], inplace=True)
df = df.drop('empty', axis = 1)
```

```python
df.head(100)
```

```python
null_data = df.isnull().sum()
null_data = null_data[null_data > 0]  
percentage = (null_data / len(df)) * 100  

print("Columns with Null Values:")
print(null_data)
print("\nPercentage of Null Values:")
print(percentage)
```

```python
columns = ['hdpe', 'ldpe', 'o', 'pet', 'pp', 'ps', 'pvc', 'grand_total', 'volunteers']

for col in columns:
    mean_value = df[col].mean()
    df[col].fillna(value=mean_value, inplace=True)

print(df.isnull().sum())
```
